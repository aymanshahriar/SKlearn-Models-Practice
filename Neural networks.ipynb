```python
# Generate an artificial dataset of 100 datapoints, containing 2 features (and obviously 1 target)
from sklearn.datasets import make_classification
x,y = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, random_state=3)

import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6), dpi=80)
x_class0 = x[y==0]
x_class1 = x[y==1]

plt.scatter(x_class0[:, 0], x_class0[:, 1], s=100, edgecolors='k')
plt.scatter(x_class1[:, 0], x_class1[:, 1], s=100, edgecolors='k', marker='v')
```

```python
# Generate and train a MLPClassifier model, which is a Multi-Layer Preceptron neural network used to solve classification problems
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification

# Create an artifical dataset containing 100 datapoints and 2 features (and obviously 1 target)
x, y = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, random_state=3)
x_train, x_test, y_train, y_test = train_test_split(x, y)

# Create and train a MLPClassifier model
MLPNeuralNetworkModel = MLPClassifier(max_iter=1000)
MLPNeuralNetworkModel.fit(x_train, y_train)

# Calculate the accuracy score of our model
print('accuracy score:', MLPNeuralNetworkModel.score(x, y))
```
